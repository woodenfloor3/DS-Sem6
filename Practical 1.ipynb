```python
#importing libraries
import pandas as pd
#create dataframe
data={'name':['Rojen','Iffat','Ayushi','Pooja','Sumit'],
      'age':[42,52,36,24,73],
      'preTestscore':[4,24,31,2,30],
      'postTestscore':[25,94,57,62,70]}
df = pd.DataFrame(data,columns =['name','age','preTestscore','postTestscore'])
df
```

```python
#print age sum
df['age'].sum()
```

```python
#print mean
df['preTestscore'].mean()
```

```python
#cummulative sum
df['preTestscore'].cumsum()
```
